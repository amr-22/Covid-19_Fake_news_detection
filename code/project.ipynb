```python
import numpy as np
import pandas as pd
import re
from tqdm import tqdm
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer

import sys
```

```python
# read csv data
data= pd.read_csv("../dataset/train/Constraint_Train.csv")
data.head(20)
```

```python
# convert dataframe(pandas) to array (numpy)
np_data=data.to_numpy()
```

```python
#split data to real and fake
real_array = []
fake_array=[]

for item in np_data:
    if(item[2]=='real'):
        real_array.append(item)
    if(item[2]=='fake'):
        fake_array.append(item)
```

```python
# return data from array (numpy) to dataframe(pandas)
real_dataframe=pd.DataFrame(real_array,columns=['id','tweet','label'])
real_dataframe
```
